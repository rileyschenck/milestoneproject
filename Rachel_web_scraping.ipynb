```python
data=pd.read_csv('Maddow_df_uniltered.csv').drop(columns='Unnamed: 0').drop_duplicates().reset_index(drop=True)
```

```python
data
```

```python
url_list = data['URL']
len(url_list)
```

```python
for n, url in enumerate(url_list):
    response = requests.get(url)
    soup = BeautifulSoup(response.content)
    results=soup.find(id='content')
    body=results.find_all('div', class_='article-body__content')
    text=body[0].text
    data['transcript'][n] = text 
```
